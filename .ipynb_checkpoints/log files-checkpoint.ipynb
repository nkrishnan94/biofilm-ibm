```python
def getLogOut(file):
    i=0
    d_str = []
    with open (file) as myfile:
        for line in myfile:

            if "Loop time" in line:
                i=0
            if i==1:
                d_str.append(line)
            if "Step Temp Press" in line:
                i=1
                
    return np.loadtxt(np.array(d_str))
```

```python
thickness = np.array([3,6,9,12])
data_LJ=np.zeros((400,8))
data_KA=np.zeros((400,8))
data_DW=np.zeros((400,8))
xTot=100
zTot=6
yTot=80
dx= 1.5
datas = [data_LJ,data_KA,data_DW]
logFiles = [LJ_logf,KA_logf,DW_logf]
outFiles = [LJ_outa,KA_outa,DW_outa]
cnt=0
for t in range(3):
    logFile =logFiles[t]
    outFile=outFiles[t]
    
    for l,a in zip(logFile,outFile):
        try:
            cnt+=1
            thicka =int(a.split('thick_')[1].split('_')[0])
            thickl =int(l.split('thick_')[1].split('_')[0])
            IDa =(int(a.split('rep_')[1].split('.')[0]))+(100*(np.argwhere(thicka==thickness)[0][0]))
            IDl =int(l.split('rep_')[1].split('.')[0])+(100*(np.argwhere(thickl==thickness)[0][0]))
            datas[t][IDa,7] = IDa
            datas[t][IDa,0] = thicka
            tTot= len(getTimeArr(a))
            cellNum = int(getAtomNumber(a))

            atoms = getLastCellArr(a,tTot,cellNum)[:-1]*np.array([xTot,yTot,zTot])
            atomsz = atoms[:,:2]
            ymax = np.zeros(len(np.arange(0,xTot,dx)))
            ymin = np.zeros(len(np.arange(0,xTot,dx)))
            for j in range(len(np.arange(0,xTot-dx,dx))):
                ymax[j]=np.max(atomsz[(atomsz[:,0]<np.arange(0,xTot,dx)[j+1]) & (atomsz[:,0]>np.arange(0,xTot,dx)[j]) ,1])
                ymin[j]=np.min(atomsz[(atomsz[:,0]<np.arange(0,xTot,dx)[j+1]) & (atomsz[:,0]>np.arange(0,xTot,dx)[j]) ,1])
            distx=0
            distn=0
            dx= np.arange(0,xTot-dx,dx)[1]-np.arange(0,xTot-dx,dx)[0]
            for j in range(len(np.arange(0,xTot-2*dx,dx))):
                distx+=np.sqrt(dx**2 +(ymax[j+1]-ymax[j])**2)
                distn +=np.sqrt(dx**2 +(ymin[j+1]-ymin[j])**2)
            x = np.arange(0,xTot-dx,dx)
            y = ymin
            #x=atomsz[:,0]
            #y=atomsz[:,1]
            regressionLineOrder = 6
            #regressionLine = np.polyfit(np.sort(x), y[np.argsort(x)],regressionLineOrder)
            regressionLine = np.polyfit(np.sort(x), y[np.argsort(x)],regressionLineOrder)
            p = np.poly1d(regressionLine)
            op = np.diff(np.sign(np.diff(p(np.sort(x))))).nonzero()[0] + 1 
            ht =np.max(ymin[:-1])- np.min(ymin[:-1])
            datas[t][IDa,1] = ht
            datas[t][IDa,2]= distx
            datas[t][IDa,3]= distn
            datas[t][IDa,4]=len(op)
            if len(op)>0:
                datas[t][IDa,5] = np.mean(np.diff(op))
            else:
                print(a)

            if t<3:
                datas[t][IDl,6]=getLogOut(l)[-1,-1]
                
        except:
            print('exception')

```
